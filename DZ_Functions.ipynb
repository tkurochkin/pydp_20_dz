```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

```

```python
#Задание 1
#Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру

def p():
    comm = input('input command: ')
    while comm != 'q':
        if comm != 'p':
            print('Неизвестная команда!')
            comm = input('input command: ')
        else:
            doc_num = input('Введите номер документа: ')
            if doc_num == 'q':
                break
            res = []
            for doc in documents:
                if doc_num in doc.values():
                        res.append(doc['name'])
            if len(res) == 0:
                print('Документ не найден в базе')
            else:
                print(f'Владелец документа: {res[0]}')


p()
```

```python
#Пункт 2. Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится

def s():
    comm = input('input command: ')
    while comm != 'q':
        if comm != 's':
            print('Неизвестная команда!')
            comm = input('input command: ')
        else:
            doc_num = input('Введите номер документа: ')
            if doc_num == 'q':
                break
            res = []
            for k, v in directories.items():
                for i in v:
                    if i == doc_num:
                        res.append(k)
            if len(res) == 0:
                print('Документ не найден в базе')
            else:
                print(f'Документ хранится на полке: {res[0]}')


s()
```

```python
#Пункт 4. Пользователь по команде "ads" может добавить новую полку
#Примеры работы:

#1.

#Введите команду:
#ads

#Введите номер полки:
#10
# Результат:
# Полка добавлена. Текущий перечень полок: 1, 2, 3, 10.

# 2.

# Введите команду:
# ads

# Введите номер полки:
# 1
# Результат:
# Такая полка уже существует. Текущий перечень полок: 1, 2, 3.




def ads():
    comm = input('Введите команду: ')
    while comm != 'q':
        if comm != 'ads':
            print('Неизвестная команда!')
            comm = input('Введите команду: ')
        else:
            sn = input('Введите номер полки: ')
            if sn == 'q':
                break
            sl =[]
            if sn in directories.keys():
                for k in directories.keys():
                    sl.append(k)
                print('Такая полка уже существует. Текущий перечень полок: ' +  ', '.join(sl))   
            if sn not in directories.keys():
                directories[sn] = []
                for k in directories.keys():
                    sl.append(k)
                print ('Полка добавлена. Текущий перечень полок: ' +  ', '.join(sl))

ads()
```

```python
# Пункт 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
# Примеры работы:

# 1.

# Введите команду:
# ds

# Введите номер полки:
# 3
# Результат:
# Полка удалена. Текущий перечень полок: 1, 2.

# 2.

# Введите команду:
# ds

# Введите номер полки:
# 1
# Результат:
# На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: 1, 2, 3.

# 3.

# Введите команду:
# ds

# Введите номер полки:
# 4
# Результат:
# Такой полки не существует. Текущий перечень полок: 1, 2, 3.


documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}


def ds():
    comm = input('Введите команду: ')
    while comm != 'q':
        if comm != 'ds':
            print('Неизвестная команда!')
            comm = input('Введите команду: ')
        else:
            sn = input('Введите номер полки: ')
            if sn == 'q':
                break
            sl =[]
            
            if sn not in directories.keys():
                for k in directories.keys():
                    sl.append(k)
                print ('Такой полки не существует. Текущий перечень полок: ' +  ', '.join(sl))
            if sn in directories.keys() and len(directories.get(sn)) == 0:
                del(directories[sn])
                for k in directories.keys():
                    sl.append(k)
                print('Полка удалена. Текущий перечень полок: ' +  ', '.join(sl))
            if sn in directories.keys() and len(directories.get(sn)) != 0:
                for k in directories.keys():
                    sl.append(k)
                print('На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: ' +  ', '.join(sl))
                
ds()








```

```python
# Пункт 2. Пользователь по команде "d" может удалить документ из данных
# Примеры работы:

# 1.

# Введите команду:
# d

# Введите номер документа:
# 10006
# Результат:

# Документ удален. 
# Текущий список документов:
# №: 2207 876234, тип: passport, владелец: Василий Гупкин, полка хранения: 1
# №: 11-2, тип: invoice, владелец: Геннадий Покемонов, полка хранения: 1
# 2.

# Введите команду:
# d

# Введите номер документа:
# 123456
# Результат:

# Документ не найден в базе. 
# Текущий список документов:
# №: 2207 876234, тип: passport, владелец: Василий Гупкин, полка хранения: 1
# №: 11-2, тип: invoice, владелец: Геннадий Покемонов, полка хранения: 1
# №: 10006, тип: insurance, владелец: Аристарх Павлов, полка хранения: 2


documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def message():
    for doc in documents:
        for i, j in directories.items():
            if doc.get('number') in j:
                print('№: ' +  doc.get('number') + ', тип: ' + doc.get('type') + ', владелец: ' + doc.get('name') + 
                                          ', полка хранения: ' + i)

def d():
    comm = input('Введите команду: ')
    while comm != 'q':
        if comm != 'd':
            print('Неизвестная команда!')
            comm = input('Введите команду: ')
        else:
            dn = input('Введите номер документа: ')
            if dn == 'q':
                break

            deleted = []
            for i in range(len(documents)-1, -1, -1):
                if documents[i]['number'] == dn:
                    deleted.append(documents[i])
                    del(documents[i])
                    print('Документ удален.')
                    print('Текущий список документов:')
                    message()
                               
            if len(deleted) == 0:
                print('Документ не найден в базе.')
                print('Текущий список документов:')
                message()    


d()


```

```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def message():
    for doc in documents:
        for i, j in directories.items():
            if doc.get('number') in j:
                print('№: ' +  doc.get('number') + ', тип: ' + doc.get('type') + ', владелец: ' + doc.get('name') + 
                                          ', полка хранения: ' + i)

def doc_check(dn):
    check = []
    for doc in documents:
        for v in doc.values():
            check.append(v)
            
    if dn not in check:
        print('Документ не найден в базе')
        print('Текущий список документов:')
        message()

    
def shelf_check(sn):
    check = []
    sl =[]           
    if sn not in directories.keys():
        for k in directories.keys():
            sl.append(k)
        print ('Такой полки не существует. Текущий перечень полок: ' +  ', '.join(sl))
        #sn = input('Введите номер полки: ')
    
    else:
        check.append(sn)        


def m():
    comm = input('Введите команду: ')
    while comm != 'q':
        if comm != 'm':
            print('Неизвестная команда!')
            comm = input('Введите команду: ')
        else:
            dn = input('Введите номер документа: ')
            if dn == 'q':
                break
            sn = input('Введите номер полки: ')
            if sn == 'q':
                break
            doc_check(dn)
            shelf_check(sn)
            check = []
            for i in directories.values():
                if dn in i:
                    check.append(i)
            if len(check) > 0:
                for k, v in directories.items():

                    if dn in v and k != sn:
                        v.remove(dn)
                    if dn in v and k == sn:
                        print('Документ уже лежит в этой полке')
                    if dn not in v and k == sn :
                        v.append(dn)
                        print('Документ перемещен')

       
                    
                    
m()
```
